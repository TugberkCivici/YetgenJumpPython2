# Problem 1

Fizikte, bir nesnenin sabit ivmeyle hareket ederken son hızını 
(veya hızını) bulmak için aşağıdaki denklem kullanılabilir:
```
vf = vi + at
burada:
vf= son hız
vi= ilk hız
a= hızlanma
t= zaman
```
İlk hız, ivme ve zaman verildiğinde, son hızı döndürecek bir fonksiyon yazın.

````python

vi = float(input("İlk hız değeri: "))
a = float(input("İvme değeri: "))
t = float(input("Zaman değeri: "))

vf = vi + a*t

print("Son hız:", vf, "m/s")
````

# Problem 3

1'den 100'e kadar olan sayılardan pisagor üçgeni oluşturanları ekrana yazdıran bir fonksiyon yazın.(a <= 100,b <= 100)


````python
def ucgenHesap():
    ucgenler=[]
    for a in range(1, 101):
        for b in range(a, 101):
            c = (a**2 + b**2) ** 0.5
            if c == int(c) and c <= 100:
                ucgenler.append((a,b,c))
                                
    return ucgenler

print(ucgenHesap())

````

# Problem 5

Bir kelimedeki ünlü ve ünsüz harfleri sayan bir fonksiyon yazınız.

````python
def harf_sayisi():
    kelime = input("Kelime girin: ")
    sesli_harfler = ['a', 'e', 'ı', 'i', 'o', 'ö', 'u']
    sesli_sayisi = 0
    sessiz_sayisi = 0
    for harf in kelime:
        if harf in sesli_harfler:
            sesli_sayisi += 1
        elif harf.isalpha():
            sessiz_sayisi += 1
    return sesli_sayisi, sessiz_sayisi
````
